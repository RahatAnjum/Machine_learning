```python
dta = dta.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dta.drop(["Properties Flysfo Estimated Timestamp","Date"],axis=1), 
                                                    dta["Properties Flysfo Estimated Timestamp"], test_size=0.20, 
                                                    random_state=8)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
accuracy = model.score(X_test, y_test)
print(accuracy*100,'%')
```
